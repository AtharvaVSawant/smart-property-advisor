```python
df_clean = df.drop(columns=['RAD', 'CHAS', 'ZN'])
```

```python
# Boxplots for all features to spot outliers
fig, axes = plt.subplots(3, 4, figsize=(15, 10))
axes = axes.ravel()

for i, col in enumerate(df_clean.columns[:-1]):  # exclude MEDV
    axes[i].boxplot(df_clean[col])
    axes[i].set_title(col)
    axes[i].set_ylabel('Value')

plt.tight_layout()
plt.show()

# Check outliers using IQR method
for col in df_clean.columns[:-1]:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df_clean[(df_clean[col] < Q1 - 1.5*IQR) | (df_clean[col] > Q3 + 1.5*IQR)]
    print(f"{col}: {len(outliers)} outliers")
```

```python
df_model = df_clean.copy()

# 1. Log transform CRIM (highly skewed)
df_model['CRIM'] = np.log1p(df_model['CRIM'])

# 2. Cap extreme outliers for other features
def cap_at_percentiles(df, col, lower=0.01, upper=0.99):
    lower_val = df[col].quantile(lower)
    upper_val = df[col].quantile(upper)
    df[col] = df[col].clip(lower_val, upper_val)
    return df

for col in ['B', 'RM', 'PTRATIO', 'LSTAT', 'DIS']:
    df_model = cap_at_percentiles(df_model, col)

# 3. Verify outliers are reduced
print("After treatment:")
for col in ['CRIM', 'B', 'RM', 'PTRATIO', 'LSTAT', 'DIS']:
    Q1 = df_model[col].quantile(0.25)
    Q3 = df_model[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df_model[(df_model[col] < Q1 - 1.5*IQR) | (df_model[col] > Q3 + 1.5*IQR)]
    print(f"{col}: {len(outliers)} outliers")
```

```python
print(f"Dataset shape: {df_clean.shape}")
print(f"Features: {df_clean.columns.tolist()}")
print(f"\nMissing values:\n{df_clean.isnull().sum()}")
print(f"\nBasic stats:\n{df_clean.describe()}")
```

```python
# Save the cleaned dataframe to CSV
df_clean.to_csv('data/boston_clean.csv', index=False)

print("✅ Cleaned data saved successfully!")
print(f"Saved shape: {df_clean.shape}")
print(f"Saved columns: {df_clean.columns.tolist()}")
```
